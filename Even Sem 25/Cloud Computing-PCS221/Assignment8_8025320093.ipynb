```python
%%writefile ./pyhdemu/sample/wc_mapper.py
#!/usr/bin/env python

from __future__ import print_function

import sys
import re

def main():
    pattern = re.compile("[A-Za-z][0-9A-Za-z]*")

    for line in sys.stdin:
        for word in pattern.findall(line): # Extract data ->  Transform -> Emit key-value
            print(f"WORD_{word.lower()}\t1")

        for ch in line.strip().lower():
          if ch.isalpha():   # optional filter
            print(f"{ch}\t1")

if __name__ == '__main__':
    main()
```

```python
from collections import defaultdict

# ── SAMPLE DATA ──────────────────────────────────────────
documents = [
    "hadoop is great hadoop is fast",
    "mapreduce runs on hadoop clusters",
    "hadoop stores data in hdfs",
]

# ── PHASE 1: MAP ─────────────────────────────────────────
def mapper(doc):
  return [(char, 1) for char in doc.lower() if char != " "]

map_output = []
for doc in documents:
    map_output.extend(mapper(doc))         #adds all elements of a list into another list

print("MAP OUTPUT:", map_output[:8], "...")     #key = word and value = 1

# ── PHASE 2: SHUFFLE & SORT ───────────────────────────────
shuffled = defaultdict(list)        #groups all values of the same key together
for key, value in map_output:
    shuffled[key].append(value)    #Collect all values belonging to the same key before reduction

print("\nAFTER SHUFFLE:", dict(list(shuffled.items())[:4]))

# ── PHASE 3: REDUCE ───────────────────────────────────────
def reducer(word, counts):
    return (word, sum(counts))

result = dict(reducer(k, v) for k, v in shuffled.items())     #For each word: Takes list of counts AND Computes total frequency

# Sort by frequency
result = dict(sorted(result.items(), key=lambda x: x[1], reverse=True))
print("\nFINAL WORD COUNT:", result)
```

```python
%%writefile wordcount.py

from mrjob.job import MRJob

class MRWordCount(MRJob):
    def mapper(self, _, line):
        for word in line.split():
            yield word.lower(), 1

    def reducer(self, key, values):
        yield key, sum(values)

if __name__ == "__main__":
    MRWordCount.run()
```

```python
%%writefile mapper.py
import sys
for line in sys.stdin:
    for word in line.strip().lower().split():
        print(f"{word}\t1")

%%writefile reducer.py
import sys
from collections import defaultdict
counts = defaultdict(int)
for line in sys.stdin:
    word, count = line.strip().split('\t')
    counts[word] += int(count)
for word, count in sorted(counts.items()):
    print(f"{word}\t{count}")
```
